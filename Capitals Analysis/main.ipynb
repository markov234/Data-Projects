```python
# Helper Functions

def download(filename, url):
    # We do not download again if the file already exists
    if os.path.exists(filename):
        return (str(filename) + " already exists!")

    r = requests.get(url)
    r.raise_for_status()
    data = r.text
    f = open(filename, "w", encoding="utf-8")
    f.write(data)
    f.close()

    return (str(filename) + " created!")
```

```python
#Q11: What is the distance between Camp Randall Stadium and the Wisconsin State Capital?

# copied from https://stackoverflow.com/questions/4913349/haversine-formula-in-python-bearing-and-distance-between-two-gps-points
def haversine(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points 
    on the earth (specified in decimal degrees)
    """
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    r = 3956 # Radius of earth in kilometers. Use 3956 for miles
    if c * r == 0:
        return "NaN"
    else:
        return c * r

haversine(-89.411893, 43.070231, -89.384113, 43.074645)
```

```python
#Q12: What is the distance between Belgium and France?

Belgium = capitals[capitals["country"] == "Belgium"]
Belgium_lon = Belgium.iloc[0].loc["longitude"]
Belgium_lat = Belgium.iloc[0].loc["latitude"]

France = capitals[capitals["country"] == "France"]
France_lon = France.iloc[0].loc["longitude"]
France_lat = France.iloc[0].loc["latitude"]

haversine(Belgium_lon, Belgium_lat, France_lon, France_lat)
```

```python
#Q13: What are the distances between Belgium, France and Spain?

def distance(country1, country2):
    country1 = capitals[capitals["country"] == country1]
    country1_lon = country1.iloc[0].loc["longitude"]
    country1_lat = country1.iloc[0].loc["latitude"]
    
    country2 = capitals[capitals["country"] == country2]
    country2_lon = country2.iloc[0].loc["longitude"]
    country2_lat = country2.iloc[0].loc["latitude"]
    
    return haversine(country1_lon, country1_lat, country2_lon, country2_lat)
    
list_countries = ["Belgium", "France", "Spain"]

list_distances = []
    
for country in list_countries:
    L = []
    L.append(distance("Belgium", country))
    L.append(distance("France", country))
    L.append(distance("Spain", country))
    list_distances.append(L)
    

DataFrame(list_distances, index = list_countries, columns = list_countries)
```

```python
#Q14: What is the distance between every pair of countries in the continent Australia?

list_countries = []

countries[countries["continent"] == "Australia"]

for country in countries[countries["continent"] == "Australia"]["country"]:
    list_countries.append(country)
    

list_distances = []

for country1 in list_countries:
    L = []
    for country2 in list_countries:
        L.append(distance(country1, country2))
    list_distances.append(L)

DataFrame(list_distances, index = list_countries, columns = list_countries)
```
